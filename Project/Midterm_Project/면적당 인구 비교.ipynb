```python
def make_column(data) :
    data = data[data['구분']=='계']
    data.drop(['기간'], axis=1)

    new = []

    for i in range(len(data['행정구역별'])) :
        sum = 0
        sum = data.iloc[i, 9] + data.iloc[i, 10] + data.iloc[i,11]
        new.append(int(sum))


    data['출산연령합계'] = new
    return data
```

```python
people_value = []
for i in range(7) :
    data = pd.read_excel('Data/면적당인구비교/'+ people[i])
    data = make_column(data)
    data = data.set_index('행정구역별')
    data = data.drop('합계')
    people_value.append(data['출산연령합계'].values)
data
```
