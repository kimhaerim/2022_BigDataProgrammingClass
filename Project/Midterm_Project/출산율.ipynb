```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
birth_value = {}
year = []

for i in range(7) :
    year.append(2014 + i)

for i in range(7) :
    data = pd.read_csv('data/출산율/' + str(year[i]) + ' 합계출산율.csv', encoding='cp949')
    birth_value[year[i]] = float(data[data['시군구별'] == '전국']['합계출산율'][0])

birth_value
```

```python
birth_value = pd.DataFrame(list(birth_value.items()), columns=['년도', '합계출산율'])
birth_value
```

```python
plt.rc('font', family='AppleGothic')
fig, ax = plt.subplots(figsize=(12,10))

file_name = '년도별 전국 합계출산율'
sns.lineplot(x='년도', y='합계출산율', data = birth_value)

for i, label in enumerate(birth_value['합계출산율']):
    plt.text(birth_value['년도'][i], birth_value['합계출산율'][i], label)
plt.title(file_name, fontsize=20)
plt.show()
```

```python
gangdong_birth_value = {}

for i in range(7) :
    data = pd.read_csv('data/출산율/' + str(year[i]) + ' 서울특별시_합계출산율.csv', encoding='cp949')
    gangdong_birth_value[year[i]] = float(data[data['시군구별'] == '강동구']['합계출산율'].values)

gangdong_birth_value
```

```python
gangdong_birth_value = pd.DataFrame(list(gangdong_birth_value.items()), columns=['년도', '합계출산율'])
gangdong_birth_value
```

```python
plt.rc('font', family='AppleGothic')
fig, ax = plt.subplots(figsize=(12,10))

file_name = '년도별 전국 합계출산율'
sns.lineplot(x='년도', y='합계출산율', data = birth_value, label='전국')
sns.lineplot(x='년도', y='합계출산율', data = gangdong_birth_value, label ='강동구')

for i, label in enumerate(birth_value['합계출산율']):
    plt.text(birth_value['년도'][i], birth_value['합계출산율'][i], label)
    
for i, label in enumerate(gangdong_birth_value['합계출산율']):
    plt.text(gangdong_birth_value['년도'][i], gangdong_birth_value['합계출산율'][i], label)
    
plt.title(file_name, fontsize=20)
plt.show()
```
