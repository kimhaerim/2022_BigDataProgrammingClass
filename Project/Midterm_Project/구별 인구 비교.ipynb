```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
people = []
year = []

for i in range(7) :
    year.append(2014 + i)

for i in range(7) :
    people.append(str(year[i]) + "서울시주민등록인구.xls")

new_column = []
people
```

```python
def make_column(data) :
    data = data[data['구분']=='계']
    data.drop(['기간'], axis=1)

    new = []

    for i in range(len(data['행정구역별'])) :
        sum = 0
        sum = data.iloc[i, 9] + data.iloc[i, 10] + data.iloc[i,11]
        new.append(int(sum))

    data['출산연령합계'] = new
    return data
```

```python
people_value = []
for i in range(7) :
    data = pd.read_excel('Data/면적당인구비교/'+ people[i])
    data = make_column(data)
    data = data.set_index('행정구역별')
    data = data.drop('합계')
    people_value.append(data['출산연령합계'].values)
data
```

```python
def make_percent(data) :
    data = data[data['구분']=='계']
    data.drop(['기간'], axis=1)

    new = []

    for i in range(len(data['행정구역별'])) :
        percent = 0
        percent = float((data.iloc[i, 9] + data.iloc[i, 10] + data.iloc[i,11])/data.iloc[:, 3].values[i] * 100)
        new.append(float(percent))

    data['출산연령비율'] = new
    return data
```

```python
data = pd.read_excel('Data/면적당인구비교/'+ people[0])
data = data[data['구분']=='한국인']
data = data[data['행정구역별']!='합계']
```
