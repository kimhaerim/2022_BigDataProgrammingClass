# 문자열의 분해

문자열을 특정 문자를 기준으로 나누고 싶을 때는 split() 함수를 사용

```python
# 공백을 기준으로 
s = 'Welcome to Python'
print(s.split())

#기준을 바꿀 수 있음
s = "2021.8.15"
print(s.split('.'))

s = "Hello, World!"
print(s.split(','))
```

# 문자열의 합성과 치환

문자열을 쪼갠 것을 다시 합치고 싶을 때는 join()함수를 사용

```python
print(','.join(['apple', 'grape', 'banana']))

# Multicharacter로 join 가능
print(', '.join(['apple', 'grape', 'banana']))
```

```python
# join과 split을 활용해 치환도 가능
print('-'.join('010.1234.5678'.split('.')))

# 다만 치환은 아래의 replace가 훨씬 편리함
print('010.1234.5678'.replace('.','-'))

s = 'Python world'
clist = list(s)
print(clist)
print(''.join(clist))
```

```python
# 약간의 트릭(?)이지만 escape code를 없애고 싶다면 아래의 방법을 쓰면 됨
a_string = 'I am\n\t iron man'
a_string
print(a_string)

# 일단 split으로 나눠주고
word_list = a_string.split() 
print(word_list)

# 공백으로 다시 join
refined_string = " ".join(word_list)
print(refined_string)
```

# 대문자와 소문자의 변환과 문자열 삭제

```python
# lower -> 소문자로 만들어줌
# upper -> 대문자로 만들어줌
s = "Hello World!"
print(s.lower())
print(s.upper())

# capitalize -> 문장의 맨 앞글자만 대문자로
s = 'apple. banana, pizza'
print(s.capitalize())
```

strip, lstrip, rstrip을 통해서 문자열의 앞뒤에 붙은 불필요한 문자를 제거 가능

```python
s = "   Hello, World!   "
# 앞뒤 모두 지울땐 strip
print(s.strip())
# 앞만 지울땐 lstrip
print(s.lstrip())
# 뒤만 지울땐 rstrip
print(s.rstrip())
```

```python
# 함수의 인자를 바꾸면 그 문자가 지워짐
s = "########this is sparta#####"
print(s.strip('#'))
print(s.lstrip('#'))
print(s.rstrip('#'))

# 이런 함수의 리턴도 문자열이므로, 이어붙여 쓸 수도 있음
print(s.strip('#').capitalize().lower().upper())
```

# 위키백과 자료 다운받기
wikipedia package는 wikipeida 문서의 자료를 조회하게 해줌

```python
import wikipedia

# 위키백과에서 조회하고 싶은 문서의 타이틀
wiki = wikipedia.page('Big data')
# 문서의 텍스트 내용을 조회
text = wiki.content
print(text)
```

# 워드 클라우드 만들기
wordcloud 패키지로 wordcloud를 만듦<br>

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 2000, height = 1500).generate(text) 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4.5)) # 기준 : 인치
plt.imshow(wordcloud) 
plt.show()
```

# 불용어 제거
불용어 : many, often같이 원치않는 다양한 단어 <br>
불용어 사전을 사용하여 제거함<br>
wordcloud패키지의 STOPWORD는 영어의 stopword를 미리 저장해둠 <br>
원하는 불용어도 추가적으로 제거할 수 있음

```python
from wordcloud import WordCloud, STOPWORDS
# set 자료형
print(STOPWORDS) 
```

```python
# 기존 STOPWORDS에 {data, information}을 추가
swords = STOPWORDS.union({'data', 'information'})

wordcloud = WordCloud(width = 2000, height = 1500,
                     stopwords=swords).generate(text)
plt.figure(figsize=(6, 4.5)) 
plt.imshow(wordcloud) 
plt.show()
```

카이사르 암호는 암호화하고자 하는 내용을 알파벳별로 일정한 거리만큼 옮겨서 다른 알파벳으로 바꾸는 방법입니다. 암호의 시초와 같죠.

예를 들어 거리가 4라고 하면 A 는 D가 되고 B는 E가 되는 형태이죠.

숫자와 문자열을 입력하면 그 거리만큼 밀어서 카이사르 암호를 만들어내는 함수 caesar_cipher(string, number) 를 만들어보세요

```python
def caesar_cipher(string, number) :
    input_string = list(string)
    for i in range(0, len(string)) :
        print(chr(ord(string[i]) + 4))
        input_string[i] = chr(ord(string[i]) + 4)
    string = "".join(input_string)
    return string
```
