# Numpy Array를 사용한 수치 계산 : Universal Functions

Numpy의 Array를 사용한 계산이 어떻게 이루어지는지 배움 <br>
Numpy Array를 사용한 계산은 구현하는 방법에 따라 속도가 다름 <br>
-> 빠르게 하는 핵심 : loop의 사용을 줄이고 계산을 벡터화 (NumPy universal functions)

<h3>The Slowness of Loops</h3>
Pyton의 기본 인터프리터 구현인 CPython은 매우 느림 <br>


```python
import numpy as np
np.random.seed(0)

def compute_reciprocals(values) :
    output = np.empty(len(values))
    for i in range(len(values)) :
        output[i] = 1.0 /values[i]
    return output

values = np.random.randint(1, 10, size=5)
compute_reciprocals(values)
```

100만개정도의 성분을 가진 list라면 몇 초 ~ 몇 분 정도가 걸림 <br>
numpy에서 개선하는 방법은?

<h3>Introducing UFuncs</h3>
Numpy는 벡터화된 연산을 통해 개선함<br>
Vector등의 선형대수 계산은 OpenBLAS, intelMKL같은 BLAS(Basic Linear Algebra Subprograms)라이브러리를 통해 빠르게 최적화 되어 있음 <br>
-> Array와 Scalar 모두 사용 가능 <br>

거의 모든 경우에 numpy array를 벡터화 하는 것이 python에서 loop를 쓰는 것보다 빠름 <br>
즉, loop를 보면 벡터화 할 수 있는지 체크하는 것이 필요함 

```python
np.show_config()
```

```python
print(compute_reciprocals(values))
print(1.0 / values)
# 결과는 같으나 속도가 매우 다름
```

<h3>Aggregates</h3>

-> reduce 사용 <br>
accumulate : 값을 누적

```python
x = np.arange(1,6)
print(np.add.reduce(x))
print(np.multiply.reduce(x))
```

```python
x = np.arange(1,6)
print(np.add.accumulate(x))
print(np.multiply.accumulate(x))
```
